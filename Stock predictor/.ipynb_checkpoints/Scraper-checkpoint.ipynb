```python
import urllib3
import certifi
import json
import time
import numpy as np
```

```python
# Scrapes data on num_items objects
def scrapeData(num_items, category):
    curritemnum = 0
    # expectedfields: number of fields we expect to see in server response
    expectedfields = 7
    urlprefix = 'https://www.ge-tracker.com/api/'
    authkey = 'Bearer b406f6f76682a9bad22106efd5334d22c28527e01069b29569d46e0b657fdd35'
    http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED', ca_certs=certifi.where())
    results = {}
    
    # Get dictionary of valid ids/names and store as iddict
    listurl = urlprefix + "items/names"
    request = http.request('GET', listurl, 
                           headers={'Accept': 'application/x.getracker.v1+json', 
                                    'Authorization': authkey})
    data = json.loads(request.data)
    iddict = {}
    for id in data:
        iddict[data[id]["item_id"]] = data[id]["name"]

    # Get data for num_items items from idlist and place in dataframe
    for key, value in iddict.items():
        if (curritemnum >= num_items or curritemnum >= len(iddict)):
            break
        else:
            if (category == "monthly"):
                graphurl = urlprefix + "graph/" + str(key) + "/month"
            elif (category == "daily"):
                graphurl = urlprefix + "graph/" + str(key) + "/day?day=5"
            else:
                print("Incorrect data category input. Request daily or monthly data.")
                break
            request = http.request('GET', graphurl, 
                                   headers={'Accept': 'application/x.getracker.v1+json', 
                                            'Authorization': authkey})
            itemdata = json.loads(request.data)
            if ("status" not in itemdata):
                itemdata = itemdata["data"]

                # Convert into np array and add into dict of item arrays
                itemarr = np.zeros(shape=(len(itemdata), expectedfields))
                r, c = 0, 0
                for entry in itemdata:
                    for k, v in entry.items():
                        if (c >= expectedfields):
                            break
                        if (c == 0):
                            # Convert milliseconds to hours
                            itemarr[r][c] = (v / 36000000)
                        else:
                            itemarr[r][c] = v
                        c += 1
                    r += 1
                    c = 0

                results[value] = itemarr
                curritemnum += 1
                time.sleep(1)
    return results
```

```python
result = scrapeData(10, "monthly")
```
